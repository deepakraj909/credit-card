```python
#print(x)
#print(y)
print(df.info())
print(df.isnull().sum())
```

```python

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.NaN ,strategy="most_frequent")
imp =imputer.fit(df)
df=imp.transform(df)
df
```

```python
df1 = pd.DataFrame(df)
df1.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
for col in df1:
    if df1[col].dtype=='object':
        df1[col]=lb.fit_transform(df1[col])
```

```python
print(df1.count())
print(df1.describe())
```

```python
#split the data into test and train
x=df1.iloc[:,:-1]
y=df1.iloc[:,-1]
print(x)
print(y)
```

```python
# import matplotlib.pyplot as plt
# plt.figure (figsize=(15,5))
# sns.heatmap(df1.corr(),annot=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
y_train
```

```python
# Import LogisticRegression
from sklearn.linear_model import LogisticRegression

lg = LogisticRegression()


print(lg.fit(x_train, y_train))
```

```python
y_pred = lg.predict(x_test)
y_pred
```

```python
# Import confusion_matrix
from sklearn.metrics import confusion_matrix,accuracy_score

confusion_matrix(y_test, y_pred)
```

```python
# Get the accuracy score of logreg model and print it
print("Accuracy of logistic regression classifier: ", accuracy_score(y_test ,y_pred))
```

```python
#Grid searching and making the model perform better

from sklearn.model_selection import GridSearchCV
param = [{'penalty' : ['l1', 'l2'],'solver': ['lbfgs','newton-cg'],'max_iter' : [100,150,200]}]
grid_model = GridSearchCV(estimator=lg, param_grid=param, scoring="accuracy",cv=6)

grid_model=grid_model.fit(x_train,y_train)
```

```python
best_accuracy=grid_model.best_score_
best_accuracy
```

```python
best_parameters=grid_model.best_params_
best_parameters
```
